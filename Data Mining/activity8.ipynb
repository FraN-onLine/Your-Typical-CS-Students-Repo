## Data Exploration

```python
# Count the number of houses with a swimming pool
houses_with_pool = dataset[dataset['PoolArea'] > 0].shape[0]
print(f"The number of houses with a swimming pool is: {houses_with_pool}")

```

```python
# 2. How many houses are duplex type?

# Count the number of houses that are duplex type
duplex_houses = dataset[dataset['BldgType'] == 'Duplex'].shape[0]
print(f"The number of duplex type houses is: {duplex_houses}")
```

```python
# Calculate the minimum and maximum lot area
min_lot_area = dataset['LotArea'].min()
max_lot_area = dataset['LotArea'].max()

print(f"The range of the houses' lot areas is: Min = {min_lot_area}, Max = {max_lot_area}")

```

```python
# 4.  What is the common zoning classification of the houses?
# Find the most common zoning classification
zoningClassification = dataset['MSZoning'].mode()[0] # why not just mode() ? or mode()[1] ? well it is because mode() returns a series of the most common values, and we want the first one.
# The mode() function returns the most common value(s) in the series. If there are multiple modes, it returns all of them.
# In this case, we are interested in the first mode, which is the most common zoning classification.
print(f"The most common zoning classification of the houses is: {zoningClassification}")
```

```python
# 5.  Are there houses located on the hillside?
# Check if there are houses located on the hillside
housesOnHillside = dataset[dataset['LandContour'] == 'HLS'].shape[0]
print(f"The number of houses located on the hillside is: {housesOnHillside}")
```

```python
# 6. What is the age of the oldest house in the area?
# Calculate the age of the oldest house
oldestAge = 2025 - dataset['YearBuilt'].min()

print(f"The age of the oldest house in the area is: {oldestAge} years")
```

```python
# 7. How many houses have a fireplace?
# Count the number of houses with at least one fireplace
withFireplace = dataset[dataset['Fireplaces'] > 0].shape[0]
print(f"The number of houses with a fireplace is: {withFireplace}")
```

```python
# 8.  Are there unfinished houses in the area?
# Check if there are unfinished houses in the area
unfinishedHouses = dataset[(dataset['HouseStyle'] == "1.5Unf") | (dataset['HouseStyle'] == "2.5Unf")].shape[0] #gets the number of houses that are unfinished.
print(f"The number of unfinished houses in the area is: {unfinishedHouses}")
```

```python
# 9.  What is the average total size (square feet) of the basement area of the houses?
# Calculate the average total size of the basement area
averageBasementSize = dataset['TotalBsmtSF'].mean()
print(f"The average total size of the basement area of the houses is: {averageBasementSize:.2f} square feet")
```

```python
# 10.  Are there houses with poor heating quality?
# Check if there are houses with poor heating quality
poorHeatingQualityHouses = dataset[dataset['HeatingQC'] == 'Po'].shape[0]
print(f"The number of houses with poor heating quality is: {poorHeatingQualityHouses}")
```
